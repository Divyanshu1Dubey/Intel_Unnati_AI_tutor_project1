```python
numerical_df = dataset.select_dtypes(include=['number'])
numerical_df.corr()
```

Exploratory Data Analysis

```python
plt.figure(figsize=(10, 6))
sns.heatmap(numerical_df.corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```

# Univariate Analysis

```python
dataset.hist(figsize=(20,10))
plt.plot()
```

```python
fig, axes = plt.subplots(2, 3, figsize=(15, 10))  
axes = axes.flatten() 
for i, col in enumerate(numerical_df.columns):
    sns.histplot(numerical_df[col], kde=True, ax=axes[i], color=np.random.rand(3,))
    axes[i].set_title(f"Distribution of {col}")
plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(2, 3, figsize=(15, 10))  
axes = axes.flatten() 
for i, col in enumerate(numerical_df.columns):
    sns.boxplot(numerical_df[col],  ax=axes[i], color=np.random.rand(3,))
    axes[i].set_title(f"Distribution of {col}")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(20,10))
gender_counts = dataset['Gender'].value_counts()
plt.title("Frequency Of Gender Distribution",fontsize=20)
sns.barplot(x=gender_counts.index, y=gender_counts.values,palette='dark')
plt.show()
```

```python
plt.figure(figsize=(20,10))
ax = sns.countplot(x='Parental_Education_Level',data=dataset,palette='Dark2')
for bar in ax.containers:
    ax.bar_label(bar,fontsize=15)
plt.title("Parent Educational Level",fontsize=30, fontweight='bold')
plt.xticks(fontsize=15)
plt.legend(['High School','College','PostGraduate'],fontsize=15)
plt.show()
```

```python
plt.figure(figsize=(20,8))
plt.pie(dataset['Earning Class'].value_counts(),autopct='%.2f')
plt.title("Earning Class",fontsize=20, fontweight='bold')
plt.legend(['Medium','Low','High'],fontsize=15)
plt.show()
```

```python
dataset.columns
```

```python
plt.figure(figsize=(20,10))
ax = sns.countplot(x='Level',data=dataset,palette='Dark2')
for bar in ax.containers:
    ax.bar_label(bar,fontsize=15)
plt.title("Student Level",fontsize=30, fontweight='bold')
plt.xticks(fontsize=15)
plt.legend(['Elementary','Middle School','Kindergarten','High School'],fontsize=15)
plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(20, 10))  
axes = axes.flatten() 
palettes = ['pastel', 'colorblind'] 

for i, (col, palette) in enumerate(zip(['Course Level', 'Material Level'], palettes)):
    sns.countplot(x=col, data=dataset, ax=axes[i], palette=palette)
    axes[i].set_title(f"Distribution of {col}", fontsize=20)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(20,8))
plt.pie(dataset['promoted'].value_counts(),autopct='%.2f')
plt.title("Student Promoted or Not",fontsize=20,fontweight='bold')
plt.legend(['Not Promoted','Promoted'])
plt.show()
```

# Bivariate Analysis

```python
plt.figure(figsize=(20,10))
sns.lineplot(x=dataset["Age"], y=dataset["IQ"])
plt.title('Age With Respect To IQ',fontsize=20)
plt.xlabel('Age',fontsize=15)
plt.ylabel('IQ',fontsize=15)
plt.show()
```

```python
plt.figure(figsize=(20,10))
sns.violinplot(x=dataset["Parental_Education_Level"], y=dataset["IQ"], palette="coolwarm")
plt.title('Parent Educational Level Corresponding Student IQ',fontsize=20)
```

```python
plt.figure(figsize=(20,10))
sns.heatmap(pd.crosstab(dataset['Earning Class'],dataset['promoted']))
plt.title("Earning class Student's Promotion Analysis",fontsize=20)
plt.show()
```

```python
plt.figure(figsize=(20,10))
sns.countplot(x=dataset["Level"], hue=dataset["promoted"])
plt.title("Student Level And Their Corresponding Promotion Analysis ",fontsize=20)
plt.show()
```

```python
plt.figure(figsize=(20,10))
sns.scatterplot(x=dataset["Previous_Scores"], y=dataset["Assesment Score"], hue=dataset['promoted'], palette="bright")
plt.title('Previous Score vs Current Assesment Score')
plt.show()
```

```python
plt.figure(figsize=(20,10))
sns.boxenplot(x=dataset["promoted"], y=dataset["Study Time"],  palette="deep")
plt.title("Student Promotion Corresponding Study Time",fontsize=20)
plt.show()
```

```python
plt.figure(figsize=(20,10))
sns.boxplot(x=dataset["promoted"], y=dataset["Attendance"],  palette="deep")
plt.title("Student Promotion Corresponding Attendance",fontsize=20)
plt.show()
```

```python
plt.figure(figsize=(20,10))
sns.scatterplot(x=dataset["IQ"], y=dataset["Study Time"], hue=dataset['promoted'], palette="bright")
plt.title('Relation Between IQ and study time',fontsize=20)
plt.show()
```

```python
plt.figure(figsize=(20,10))
sns.violinplot(y=dataset["Assesment Score"], x=dataset["Course Level"], palette="coolwarm")
plt.title('Relation between assesment score and study time',fontsize=20)
```

```python
plt.figure(figsize=(20, 10))
sns.countplot(x='Course Level', hue='Material Level', data=dataset)
plt.xlabel("Course Level")
plt.ylabel("Count")
plt.title(" Relation of Course Level and Material Level",fontsize=20)
plt.legend(title="Material Level", loc="upper right")
plt.show()
```

# Multivariate Analysis

```python
plt.figure(figsize=(20,10))
sns.violinplot(x='Gender',y='IQ',data=dataset,hue='promoted')
plt.title("Determining IQ of different genders and corresponding Promotion",fontsize=20)
plt.show()
```

```python
plt.figure(figsize=(20,10))
sns.heatmap(pd.crosstab(dataset['Earning Class'],dataset['Parental_Education_Level']))
plt.title("Finding relation between Earning class Student Parent Educational Level",fontsize=20)
plt.show()
```

```python
plt.figure(figsize=(20,10))
sns.boxplot(x='Gender',y='IQ',data=dataset,hue='Level')
plt.title("Gender With respect to IQ and Student Level",fontsize=20)
plt.show()
```

```python
plt.figure(figsize=(20,10))
sns.barplot(x='promoted',y='Attendance',data=dataset,hue='Level')
plt.title("Promotion With respect to Attendance and Student Level",fontsize=20)
plt.show()
```
