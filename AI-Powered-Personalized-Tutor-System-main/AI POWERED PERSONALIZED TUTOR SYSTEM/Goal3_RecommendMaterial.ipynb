```python
dataset = pd.read_csv(r"k-12.csv")
dataset.sample(10)
```

```python
dataset = dataset.drop(columns=['Name','Gender','Country','State','City','Parental_Education_Level','Earning Class','Course Name','Material Name','promoted'])
```

```python
x = dataset.drop(columns=['recommended_material'])
y = dataset['recommended_material']
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
x
```

```python
lb = LabelEncoder()

y_train = lb.fit_transform(y_train)
y_test = lb.transform(y_test)
```

```python
transformer1 = ColumnTransformer(
    transformers=[
        ('tf1', OrdinalEncoder(categories=[
              # Categories for 'Earning Class'
            ['Kindergarten', 'Elementary', 'Middle School', 'High School'],  # Categories for 'Level'
            ['Low', 'Medium', 'High'],  # Categories for 'Course Level'
            ['Low', 'Medium', 'High']  # Categories for 'Material Level'
        ]), [ 'Level', 'Course Level', 'Material Level'])  # Ordinal columns
    ],
    remainder='passthrough'
)

pipeline = Pipeline(
    [
        ('Preprocess',transformer1),
        ('Scaling',StandardScaler()),
        ('SoftMax',LogisticRegression(multi_class='multinomial'))
    ]
)
```

```python
pipeline.fit(x_train,y_train)
```

```python
y_pred = pipeline.predict(x_test)
```

```python
accuracy_score(y_test,y_pred)
```

```python
confusion_matrix(y_test,y_pred)
```

```python
# Recommend material as per student level
input_data = pd.DataFrame([[
    6,	'Kindergarten',	'Medium',	'Medium',	89,	69,	102.15,	86,	4.14
]], columns=x_train.columns)

# Make prediction
prediction = pipeline.predict(input_data)

def recommend(n):
    if n==0:
        print(y[0])
    elif n==1:
        print(y[1])
    elif n==2:
        print(y[3])
    else:
        print(y[2])            

print("Kindly Follow These Resources : ")        
recommend(prediction[0])
```
